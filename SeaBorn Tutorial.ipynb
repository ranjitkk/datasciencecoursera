```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',estimator=np.median,data=tips_df)
```

### Countplot

```python
sns.countplot(x='sex',data=tips_df)
```

# BoxPlot

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=tips_df,fliersize=5,dodge=True)
plt.legend(loc=1)
```

```python
### Violinplot
```

```python
sns.violinplot(x='day',y="tip",data=tips_df,hue='sex',split=True,dodge=True)
```

```python
sns.violinplot(x='day',y="tip",data=tips_df,hue='sex',split=False)
```

### stripplot 

```python
sns.stripplot(x='day',y="tip",hue='time',data=tips_df)
```

```python
sns.stripplot(x='day',y="tip",hue='time',data=tips_df,jitter=True,dodge=True)
```

### swarmplot

```python
sns.violinplot(x='day',y="tip",data=tips_df,hue='time',split=False,dodge=True)
sns.swarmplot(x='day',y="tip",hue='time',data=tips_df,dodge=True,color='white')
```

## Palletes

```python
plt.figure(figsize=(18,10))
sns.set_style(style='darkgrid')
sns.set_context(context='talk')
sns.stripplot(x='day',y="tip",hue='time',data=tips_df,jitter=True,dodge=True,size=10,palette='Accent')
plt.legend(loc=1)
```

### Pair Grids

```python
iris=sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')
iris_g.map_diag(plt.hist)
#iris_g.map_offdiag(sns.stripplot)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
iris_g.add_legend(title='Flowers Type')
```

```python
iris_g = sns.PairGrid(iris,x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
iris_g.map(plt.scatter)
```

### FAcetGrid

```python
tips_fg = sns.FacetGrid(tips_df,col='time',row='sex')
tips_fg.map(sns.scatterplot,'total_bill','tip')
```

```python
tips_fg = sns.FacetGrid(tips_df,col='time',hue='sex', palette='Set1')
tips_fg.map(plt.scatter,'total_bill','tip')
tips_fg.add_legend()
```

```python
att_df = sns.load_dataset('attention')
att_df.head()
```

```python
att_fg = sns.FacetGrid(att_df,col='attention',row='subject', palette='Greens')
att_fg.map(plt.plot,'solutions','score', marker='v')
```

## Regression Plots

```python
plt.figure(figsize=(20,8))
sns.set_context(context='paper')
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips_df, markers=['o','^'],scatter_kws={'s':50, 'linewidth':0.5,
                                                                                        'edgecolor':'white'})
```

```python
sns.lmplot(x='total_bill',y='tip', col='sex',hudata=tips_df)
```
